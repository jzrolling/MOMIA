```python
import numpy as np
import pickle
import pandas as pd
import glob, os
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec as gs
```

```python
# load data
data = np.load('/Users/jz-rolling/Desktop/MSR_OMEGA_manuscript/Another_try/revision_1/packed_rep1_spikein.npy')
lengths = np.load('/Users/jz-rolling/Desktop/MSR_OMEGA_manuscript/Another_try/revision_1/packed_rep1_spikein_lengths.npy')
df = pd.read_excel('/Users/jz-rolling/Desktop/MSR_OMEGA_manuscript/Another_try/revision_1/rep1_spikein_curated.xls')
print(data.shape)
# n samples, m lengths bins, 450 (15*30) features
```

```python
# reshape data:
def min_max_normalization(data, pad=0.01):
    # Min-max normalization function
    # constrain zero entries by padding the data with a 0.01 background
    return ((data-data.min())/(data.max()-data.min())+pad)/(1+pad)

def feature_plot(basis_image,vmax=None):
    from matplotlib.gridspec import GridSpec as gs
    nfeatures = basis_image.shape[0]
    ncols = 5
    if nfeatures%ncols == 0:
        nrows = int(nfeatures/ncols)
    else:
        nrows = int(nfeatures/ncols)+1
    fig = plt.figure(figsize=(ncols*2,nrows*1.2))
    grids = gs(nrows,ncols,hspace=0.4,wspace=0.1)
    for r in range(nrows):
        for c in range(ncols):
            idx = r*ncols+c
            if  idx <= nfeatures:
                ax = fig.add_subplot(grids[r,c])
                if vmax == None:
                    vmax = basis_image[idx].max()
                ax.imshow(basis_image[idx].reshape(15,30),aspect='auto')
                ax.set_xticks([])
                ax.set_yticks([])
                ax.set_title('feature {}'.format(idx+1))
    return fig

"""
L=10
stacked = np.zeros((data.shape[0]*data.shape[1],data.shape[2]))
for i in range(data.shape[0]):
    temp = min_max_normalization(data[i])
    stacked[i*L:(i+1)*L] = temp
"""

# omit PP1 dropout
curated_id = df['Original_index'].values

L=10
stacked = np.zeros((len(curated_id)*data.shape[1],data.shape[2]))
for k,i in enumerate(curated_id):
    temp = min_max_normalization(data[i])
    stacked[k*L:(k+1)*L] = temp
```

```python
stacked.shape
```

```python
# NMF analysis
# applied a relatively high tolerance to avoid overfitting
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA,NMF

n_components = 20
model = NMF(n_components=n_components,init='nndsvdar',tol=0.005,
            random_state=42,alpha=0.02,max_iter=500)
w = model.fit_transform(stacked)
h = model.components_
reshaped_features = np.zeros((len(curated_id),data.shape[1],n_components))
for i in range(len(curated_id)):
    reshaped_features[i,:,:]=w[i*L:(i+1)*L]
```

```python
# visualize basis images
fig=feature_plot(h)
```

```python
# temp resolved correlation matrix
# ignore warnings for now (zeros found in encoding data)
corr_M = np.zeros((len(curated_id),len(curated_id),data.shape[1]))
for i in range(data.shape[1]):
    corr_M[:,:,i] = np.corrcoef(reshaped_features[:,i,:])
```

```python
# LRFP plots

# helper func
def plot_temporal_dynamics(tensor, filename = None, vmin=0.5, vmax=4):
    fig = plt.figure(figsize=(2,10))
    grids = gs(10,2,hspace=0.05)
    for i in range(10):
        ax = fig.add_subplot(grids[i,:])
        plt.axis('off')
        ax.imshow(tensor[i,:,:],aspect='auto',vmin=vmin,vmax=vmax)
    if filename != None:
        plt.savefig(filename,dpi=160,bbox_inches='tight',transparent=True)
        plt.close()
```
